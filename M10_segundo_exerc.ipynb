````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df = pd.read_csv(r"C:\Users\Adm\Desktop\EBAC_ex\previsao_de_renda.csv")
````

````python
fig, axes = plt.subplots(1, 2,figsize=(8, 4), gridspec_kw=dict(width_ratios=[4, 4]) )

sns.pointplot(x='posse_de_veiculo', y='renda', data=df, ax=axes[0])
sns.pointplot(x='posse_de_imovel', y='renda', data=df,color='tab:orange',ax=axes[1])
fig.tight_layout();
````

Ambas têm diferença estatisticamente significantes, logo, são boas preditoras de renda.
