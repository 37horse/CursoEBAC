1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df = pd.read_csv(r"C:\Users\Adm\Desktop\EBAC_ex\previsao_de_renda.csv")
````

````python
var_interesse = ['sexo','posse_de_veiculo','posse_de_imovel','tipo_renda',
                 'educacao','estado_civil','tipo_residencia', 'renda']
len(var_interesse) - 2
````

````python
fig, axes = plt.subplots(3,2,figsize=(12,15))

df['data_ref_m'] = list(map(lambda x: x[0:7], df['data_ref']))
labels = df['data_ref_m'].unique()

i = 0
for column in range(0,2):
    for row in range(0,3):
        sns.countplot(x='data_ref_m',hue=var_interesse[i],data=df, ax=axes[row,column]).set_xticklabels(labels=labels,
                                                                                                        rotation=60)
        i = i+1

fig.tight_layout()
````

Todas as variáveis têm uma variação aceitável para se considerar estabilidade, com execeção da variável educação, que teve um acréscimo de ensino superior completo e uma diminuição para o ensino secundário a partir de setembro de 2015.

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
fig, axes = plt.subplots(3,2,figsize=(12,15))

i=0
for row in range(0,3):
    for column in range(0,2):
        sns.pointplot(x='data_ref_m', y='renda', hue=var_interesse[i], data=df, dodge=True, ci=95, ax=axes[row,column])\
        .set_xticklabels(labels=labels,rotation=60);
        i+=1

fig.tight_layout()
````

As variáveis mais estáveis são: Estado Civil, Posse de Imóvel, Posse de Veículo e Sexo.
